```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mimg
```

```python
#NOTE : here we have created arrays using different methods
array1 = np.array([[1,2,3],[5,1,5]])
array2 = np.random.rand(3,3)    # this creates 3X3 matrice
array3 = np.ones((3,3)) # this matrice will contain only element - 1
```

### Saving and loading arrays of numpy

```python
#NOTE : here we saved above arrays in the same directory 
#NOTE : syntax for saving :    np.save('path', arrayname)

np.save('array1.npy', array1)
np.save('array2.npy', array2)
np.save('array3.npy', array3)
```

```python
#NOTE : here we loaded above saved arrays 
#NOTE : syntax for loading :        np.load('file_name.npy')

loaded_array1 = np.load('array1.npy')
loaded_array2 = np.load('array2.npy')
loaded_array3 = np.load('array3.npy')

print("Loaded array 1 : ", loaded_array1)
print("\n")
print("Loaded array 2 : ", loaded_array2)
print("\n")
print("Loaded array 3 : ", loaded_array3)
print("\n")
```

### NOW CHANGING THE COLOR OF THE LOGO OF Real Madrid 

```python
#NOTE : mpimg.imread() function reads the image and returns a Numpy array

try:
    loading_logo_matrix = mimg.imread('RealMadrid.png')
    plt.figure(figsize=(10, 5))     # usually this fits of (10, 6)
    plt.subplot(121)
    plt.imshow(loading_logo_matrix)
    plt.grid(False)
    plt.title("Real Madrid")


    #NOTE : now making darkmode 
    dark_mode = 1 - loading_logo_matrix     # so this is trick by finding 1's complement
    plt.subplot(122)
    plt.imshow(dark_mode)
    plt.grid(False)
    plt.title("Real Madrid")
    
except FileExistsError:
    print("Numpy logo file not found !")
```
